```python
import os, sys, math, time, random
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score, adjusted_rand_score, normalized_mutual_info_score
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

from google.colab import drive
drive.mount('/content/drive', force_remount=False)

try:
    import umap
    UMAP_AVAILABLE = True
except Exception:
    UMAP_AVAILABLE = False
    print("umap not available — UMAP visualizations will fall back to t-SNE (slower).")

# Config
CSV_PATH = "/content/drive/MyDrive/CSE425/cybersecurity_intrusion_data.csv"
OUTPUT_DIR = "/content/drive/MyDrive/CSE425/vade_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

USE_LOG1P = True
LOG1P_FEATURES = ["session_duration", "network_packet_size"]
SCALER = "robust"
PCA_COMPONENTS = None
LATENT_DIM = 20
HIDDEN_DIMS = [256, 128]
N_CLUSTERS_TRY = [2, 3]
BATCH_SIZE = 128
EPOCHS_VAE = 60
EPOCHS_AE = 60
EPOCHS_DEC = 40
SEEDS = [0, 1, 2, 3]
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
PRINT_EVERY = 10

# Utility functions
def set_seed(seed=0):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

def ensure_torch(x, dtype=torch.float32, device=DEVICE):
    if isinstance(x, np.ndarray):
        return torch.from_numpy(x).to(dtype=dtype, device=device)
    if isinstance(x, torch.Tensor):
        return x.to(dtype=dtype, device=device)
    raise ValueError("Unsupported type")

# Data load + preprocess
def load_and_preprocess(csv_path,
                        use_log1p=USE_LOG1P,
                        log1p_features=LOG1P_FEATURES,
                        scaler_type=SCALER,
                        pca_components=PCA_COMPONENTS):
    df = pd.read_csv(csv_path)
    print("Original CSV shape:", df.shape)
    if 'session_id' in df.columns:
        df = df.drop(columns=['session_id'])
    y = None
    if 'attack_detected' in df.columns:
        y = df['attack_detected'].astype(int).values
        df = df.drop(columns=['attack_detected'])

    for col in log1p_features:
        if col in df.columns and np.issubdtype(df[col].dtype, np.number) and use_log1p:
            print(f"Applying log1p to {col}")
            df[col] = np.log1p(df[col].values)

    categorical_cols = [c for c in df.columns if df[c].dtype == 'object']
    numeric_cols = [c for c in df.columns if c not in categorical_cols]
    print("Categorical columns:", categorical_cols)
    print("Numeric columns:", numeric_cols)

    if categorical_cols:
        df = pd.get_dummies(df, columns=categorical_cols, drop_first=False)

    X = df.values.astype(np.float32)
    if scaler_type == "robust":
        scaler = RobustScaler()
    else:
        scaler = StandardScaler()
    X = scaler.fit_transform(X)

    if pca_components is not None and pca_components < X.shape[1]:
        print(f"Applying PCA (n_components={pca_components})")
        pca = PCA(n_components=pca_components, random_state=0)
        X = pca.fit_transform(X)

    return X, y, scaler, df.columns.tolist()

# Dataloaders
def make_dataloaders(X, batch_size=BATCH_SIZE, val_frac=0.1):
    X_train, X_val = train_test_split(X, test_size=val_frac, random_state=42)
    train_loader = DataLoader(TensorDataset(torch.from_numpy(X_train)), batch_size=batch_size, shuffle=True, drop_last=True)
    val_loader = DataLoader(TensorDataset(torch.from_numpy(X_val)), batch_size=batch_size, shuffle=False)
    return train_loader, val_loader, X_train, X_val

# Models
class Encoder(nn.Module):
    def __init__(self, input_dim, hidden_dims, latent_dim, dropout=0.1):
        super().__init__()
        layers = []
        last = input_dim
        for h in hidden_dims:
            layers += [nn.Linear(last, h), nn.ReLU(), nn.BatchNorm1d(h), nn.Dropout(dropout)]
            last = h
        self.net = nn.Sequential(*layers)
        self.fc_mu = nn.Linear(last, latent_dim)
        self.fc_logvar = nn.Linear(last, latent_dim)
    def forward(self, x):
        h = self.net(x)
        return self.fc_mu(h), self.fc_logvar(h)

class Decoder(nn.Module):
    def __init__(self, latent_dim, hidden_dims, output_dim, dropout=0.1):
        super().__init__()
        layers = []
        last = latent_dim
        for h in hidden_dims[::-1]:
            layers += [nn.Linear(last, h), nn.ReLU(), nn.BatchNorm1d(h), nn.Dropout(dropout)]
            last = h
        layers.append(nn.Linear(last, output_dim))
        self.net = nn.Sequential(*layers)
    def forward(self, z):
        return self.net(z)

class VAE_GMM(nn.Module):
    def __init__(self, input_dim, latent_dim=LATENT_DIM, hidden_dims=HIDDEN_DIMS, n_clusters=2, dropout=0.1):
        super().__init__()
        self.encoder = Encoder(input_dim, hidden_dims, latent_dim, dropout)
        self.decoder = Decoder(latent_dim, hidden_dims, input_dim, dropout)
        self.latent_dim = latent_dim
        self.n_clusters = n_clusters
        self.pi_logits = nn.Parameter(torch.zeros(n_clusters))
        self.cluster_mu = nn.Parameter(torch.randn(n_clusters, latent_dim) * 0.01)
        self.cluster_logvar = nn.Parameter(torch.zeros(n_clusters, latent_dim))
    def reparameterize(self, mu, logvar):
        eps = torch.randn_like(mu)
        return mu + eps * torch.exp(0.5 * logvar)
    def forward(self, x):
        mu, logvar = self.encoder(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decoder(z)
        return recon, mu, logvar, z
    @staticmethod
    def log_normal_diag(x, mu, logvar):
        return -0.5 * (torch.log(2 * torch.tensor(np.pi, device=x.device)) + logvar + ((x - mu)**2)/torch.exp(logvar)).sum(dim=-1)
    def compute_loss(self, x, recon, mu, logvar, z, beta=1.0):
        recon_loss = F.mse_loss(recon, x, reduction='sum') / x.size(0)
        log_qzx = VAE_GMM.log_normal_diag(z, mu, logvar)
        B = z.size(0); K = self.n_clusters; D = self.latent_dim
        z_exp = z.unsqueeze(1).expand(B, K, D)
        mu_k = self.cluster_mu.unsqueeze(0).expand(B, K, D)
        logvar_k = self.cluster_logvar.unsqueeze(0).expand(B, K, D)
        log_pzk = -0.5 * (torch.log(2 * torch.tensor(np.pi, device=z.device)) + logvar_k + ((z_exp - mu_k)**2)/torch.exp(logvar_k)).sum(-1)
        log_pi = F.log_softmax(self.pi_logits, dim=0)
        log_probs = log_pi.unsqueeze(0) + log_pzk
        log_pz = torch.logsumexp(log_probs, dim=1)
        kl = (log_qzx - log_pz).mean()
        loss = recon_loss + beta * kl
        return loss, float(recon_loss.item()), float(kl.item())
    def predict_cluster(self, z):
        if isinstance(z, np.ndarray):
            z = torch.from_numpy(z).to(next(self.parameters()).device)
        z = z.to(next(self.parameters()).device)
        B = z.size(0); K = self.n_clusters; D = self.latent_dim
        z_exp = z.unsqueeze(1).expand(B, K, D)
        mu_k = self.cluster_mu.unsqueeze(0).expand(B, K, D)
        logvar_k = self.cluster_logvar.unsqueeze(0).expand(B, K, D)
        log_pzk = -0.5 * (torch.log(2 * torch.tensor(np.pi, device=z.device)) + logvar_k + ((z_exp - mu_k)**2)/torch.exp(logvar_k)).sum(-1)
        log_pi = F.log_softmax(self.pi_logits, dim=0)
        log_probs = log_pi.unsqueeze(0) + log_pzk
        probs = F.softmax(log_probs, dim=1).detach().cpu().numpy()
        labels = probs.argmax(axis=1)
        return labels, probs

class DeterministicAE(nn.Module):
    def __init__(self, input_dim, latent_dim=LATENT_DIM, hidden_dims=HIDDEN_DIMS, dropout=0.1):
        super().__init__()
        layers = []; last = input_dim
        for h in hidden_dims:
            layers += [nn.Linear(last, h), nn.ReLU(), nn.BatchNorm1d(h), nn.Dropout(dropout)]
            last = h
        self.encoder = nn.Sequential(*layers, nn.Linear(last, latent_dim))
        layers = []; last = latent_dim
        for h in hidden_dims[::-1]:
            layers += [nn.Linear(last, h), nn.ReLU(), nn.BatchNorm1d(h), nn.Dropout(dropout)]
            last = h
        layers.append(nn.Linear(last, input_dim))
        self.decoder = nn.Sequential(*layers)
    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z), z

class DECClustering(nn.Module):
    def __init__(self, n_clusters, latent_dim, alpha=1.0):
        super().__init__()
        self.n_clusters = n_clusters
        self.latent_dim = latent_dim
        self.alpha = alpha
        self.cluster_centers = nn.Parameter(torch.randn(n_clusters, latent_dim))
    def forward(self, z):
        z_exp = z.unsqueeze(1)
        centers = self.cluster_centers.unsqueeze(0)
        dist_sq = torch.sum((z_exp - centers)**2, dim=2)
        q = 1.0 / (1.0 + (dist_sq / self.alpha))
        q = q ** ((self.alpha + 1.0) / 2.0)
        q = (q.t() / torch.sum(q, dim=1)).t()
        return q

def target_distribution(q):
    if isinstance(q, torch.Tensor):
        q_np = q.detach().cpu().numpy()
    else:
        q_np = q
    f = (q_np ** 2) / np.sum(q_np, axis=0, keepdims=True)
    p = f / np.sum(f, axis=1, keepdims=True)
    return torch.from_numpy(p).float()

# Training routines
def train_vade(model, train_loader, val_loader, epochs=EPOCHS_VAE, lr=1e-3, weight_decay=1e-5, patience=10, beta=1.0, save_path="vade.pth"):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    best_val = np.inf; ctr = 0
    model.to(DEVICE)
    history = {"train_loss": [], "val_loss": []}
    for epoch in range(1, epochs+1):
        model.train(); train_losses=[]
        for (xb,) in train_loader:
            xb = xb.to(DEVICE).float()
            optimizer.zero_grad()
            recon, mu, logvar, z = model(xb)
            loss, rec_l, kl = model.compute_loss(xb, recon, mu, logvar, z, beta=beta)
            loss.backward(); optimizer.step()
            train_losses.append(loss.item())

        model.eval(); val_losses=[]
        with torch.no_grad():
            for (xb,) in val_loader:
                xb = xb.to(DEVICE).float()
                recon, mu, logvar, z = model(xb)
                loss, rec_l, kl = model.compute_loss(xb, recon, mu, logvar, z, beta=beta)
                val_losses.append(loss.item())
        train_loss = np.mean(train_losses); val_loss = np.mean(val_losses)
        history["train_loss"].append(train_loss); history["val_loss"].append(val_loss)
        if epoch % PRINT_EVERY == 0 or epoch == 1:
            print(f"[VaDE] Epoch {epoch}/{epochs} train_loss={train_loss:.4f} val_loss={val_loss:.4f}")
        if val_loss < best_val - 1e-5:
            best_val = val_loss; ctr = 0; torch.save(model.state_dict(), save_path)
        else:
            ctr += 1
            if ctr >= patience:
                print("VaDE early stopping at epoch", epoch)
                break
    model.load_state_dict(torch.load(save_path, map_location=DEVICE))
    return model, history

def train_ae(ae, train_loader, val_loader, epochs=EPOCHS_AE, lr=1e-3, weight_decay=1e-5, patience=10, save_path="ae.pth"):
    optimizer = torch.optim.Adam(ae.parameters(), lr=lr, weight_decay=weight_decay)
    best_val = np.inf; ctr = 0
    ae.to(DEVICE)
    for epoch in range(1, epochs+1):
        ae.train(); train_losses=[]
        for (xb,) in train_loader:
            xb = xb.to(DEVICE).float()
            optimizer.zero_grad()
            recon, z = ae(xb)
            loss = F.mse_loss(recon, xb, reduction='mean')
            loss.backward(); optimizer.step()
            train_losses.append(loss.item())
        ae.eval(); val_losses=[]
        with torch.no_grad():
            for (xb,) in val_loader:
                xb = xb.to(DEVICE).float()
                recon, z = ae(xb)
                val_losses.append(F.mse_loss(recon, xb, reduction='mean').item())
        train_loss = np.mean(train_losses); val_loss = np.mean(val_losses)
        if epoch % PRINT_EVERY == 0 or epoch == 1:
            print(f"[AE] Epoch {epoch}/{epochs} train={train_loss:.5f} val={val_loss:.5f}")
        if val_loss < best_val - 1e-6:
            best_val = val_loss; ctr = 0; torch.save(ae.state_dict(), save_path)
        else:
            ctr += 1
            if ctr >= patience:
                print("AE early stopping at epoch", epoch); break
    ae.load_state_dict(torch.load(save_path, map_location=DEVICE))
    return ae

def train_dec(ae, X, n_clusters, epochs=EPOCHS_DEC, tol=1e-3, lr=1e-3):
    """
    DEC training: refine AE encoder + cluster centers.
    Returns updated AE and DEC clustering module.
    """
    ae_local = ae
    ae_local.to(DEVICE); ae_local.eval()
    # compute initial latent for all data
    with torch.no_grad():
        z_all = ae_local.encoder(torch.from_numpy(X).float().to(DEVICE)).cpu().numpy()
    # init kmeans
    kmeans = KMeans(n_clusters=n_clusters, n_init=20, random_state=0).fit(z_all)
    cluster_centers = torch.tensor(kmeans.cluster_centers_, dtype=torch.float32).to(DEVICE)
    clustering = DECClustering(n_clusters=n_clusters, latent_dim=z_all.shape[1]).to(DEVICE)
    clustering.cluster_centers.data = cluster_centers.clone().detach()
    # optimizer: encoder params + cluster centers
    optimizer = torch.optim.Adam(list(ae_local.encoder.parameters()) + list(clustering.parameters()), lr=lr)
    X_t = torch.from_numpy(X).float().to(DEVICE)
    prev_q = None
    for epoch in range(1, epochs+1):
        ae_local.train()
        q_list = []
        batch_size = 256
        for i in range(0, X.shape[0], batch_size):
            xb = X_t[i:i+batch_size]
            optimizer.zero_grad()
            z_batch = ae_local.encoder(xb)
            q_batch = clustering(z_batch)
            p_batch = target_distribution(q_batch).to(DEVICE)
            loss = F.kl_div(q_batch.log(), p_batch, reduction='batchmean')
            loss.backward()
            optimizer.step()
            q_list.append(q_batch.detach().cpu().numpy())
        q_all = np.vstack(q_list)
        if prev_q is None:
            delta = None
        else:
            delta = float(np.mean(np.abs(q_all - prev_q)))
        prev_q = q_all
        if epoch % 10 == 0 or epoch == 1:
            print(f"[DEC] epoch {epoch}/{epochs} delta {0.0 if delta is None else delta:.6f}")
        if delta is not None and delta < tol:
            print("DEC converged (delta < tol) at epoch", epoch)
            break
    return ae_local, clustering

# Evaluation and visualization helpers
def evaluate_clustering(latent, pred_labels, true_labels):
    metrics = {}
    try:
        metrics['silhouette'] = float(silhouette_score(latent, pred_labels))
    except Exception:
        metrics['silhouette'] = float('nan')
    metrics['ARI'] = float(adjusted_rand_score(true_labels, pred_labels)) if true_labels is not None else float('nan')
    metrics['NMI'] = float(normalized_mutual_info_score(true_labels, pred_labels)) if true_labels is not None else float('nan')
    return metrics

def predictive_uncertainty(model, X_np, n_samples=30):
    model.to(DEVICE); model.eval()
    with torch.no_grad():
        X_t = torch.from_numpy(X_np).float().to(DEVICE)
        mu, logvar = model.encoder(X_t)
        N = X_np.shape[0]; K = model.n_clusters
        probs_acc = np.zeros((N, K), dtype=np.float32)
        for i in range(n_samples):
            z = model.reparameterize(mu, logvar)
            _, probs = model.predict_cluster(z)
            probs_acc += probs
        probs_mean = probs_acc / n_samples
        entropy = -np.sum(probs_mean * np.log(probs_mean + 1e-12), axis=1)
        return probs_mean, entropy

def plot_umap(z, labels, title, fname):
    if UMAP_AVAILABLE:
        reducer = umap.UMAP(n_components=2, random_state=0)
        emb = reducer.fit_transform(z)
    else:
        emb = TSNE(n_components=2, random_state=0).fit_transform(z)
    plt.figure(figsize=(6,5))
    sns.scatterplot(x=emb[:,0], y=emb[:,1], hue=labels, palette='tab10', s=8, legend='full')
    plt.title(title)
    plt.tight_layout(); plt.savefig(fname, dpi=150)
    plt.show()
    plt.close()

def plot_tsne(z, labels, title, fname):
    emb = TSNE(n_components=2, perplexity=30, random_state=0).fit_transform(z)
    plt.figure(figsize=(6,5))
    sns.scatterplot(x=emb[:,0], y=emb[:,1], hue=labels, palette='tab10', s=8, legend='full')
    plt.title(title)
    plt.tight_layout(); plt.savefig(fname, dpi=150)
    plt.show()
    plt.close()

def plot_entropy_hist(entropy, fname):
    plt.figure(figsize=(6,3))
    sns.histplot(entropy, bins=50, kde=True)
    plt.title("Predictive entropy distribution")
    plt.tight_layout(); plt.savefig(fname, dpi=150)
    plt.show()
    plt.close()

# Full run orchestrator
def run_all(csv_path=CSV_PATH,
            output_dir=OUTPUT_DIR,
            try_n_clusters=N_CLUSTERS_TRY,
            latent_dim=LATENT_DIM,
            pca_components=PCA_COMPONENTS,
            seeds=SEEDS):
    X, y, scaler, colnames = load_and_preprocess(csv_path, pca_components=pca_components)
    n, dim = X.shape
    print("Processed X shape:", X.shape)
    if y is not None:
        print("Label counts:", Counter(y))
    results_summary = {}

    for n_clusters in try_n_clusters:
        print("\n" + "="*40)
        print("n_clusters =", n_clusters)
        vade_metrics=[]; ae_metrics=[]; gmm_metrics=[]; dec_metrics=[]
        for seed in seeds:
            print("--- seed", seed)
            set_seed(seed)
            train_loader, val_loader, X_train, X_val = make_dataloaders(X)
            vade = VAE_GMM(input_dim=dim, latent_dim=latent_dim, hidden_dims=HIDDEN_DIMS, n_clusters=n_clusters)
            vade, _ = train_vade(vade, train_loader, val_loader, epochs=EPOCHS_VAE, save_path=os.path.join(output_dir, f"vade_k{n_clusters}_s{seed}.pth"))
            vade.eval()
            with torch.no_grad():
                X_t = torch.from_numpy(X).float().to(DEVICE)
                mu_all, _ = vade.encoder(X_t)
                z_mu = mu_all.cpu().numpy()
            labels_vade, probs_vade = vade.predict_cluster(z_mu)
            vade_metrics.append(evaluate_clustering(z_mu, labels_vade, y))
            probs_mean, entropy = predictive_uncertainty(vade, X, n_samples=30)
            np.save(os.path.join(output_dir, f"vade_probs_k{n_clusters}_s{seed}.npy"), probs_mean)
            np.save(os.path.join(output_dir, f"vade_entropy_k{n_clusters}_s{seed}.npy"), entropy)
            plot_entropy_hist(entropy, os.path.join(output_dir, f"entropy_k{n_clusters}_s{seed}.png"))

            # AE + KMeans baseline
            ae = DeterministicAE(input_dim=dim, latent_dim=latent_dim, hidden_dims=HIDDEN_DIMS)
            ae = train_ae(ae, train_loader, val_loader, epochs=EPOCHS_AE, save_path=os.path.join(output_dir, f"ae_k{n_clusters}_s{seed}.pth"))
            ae.eval()
            with torch.no_grad():
                X_t = torch.from_numpy(X).float().to(DEVICE)
                _, z_ae = ae(X_t)
                z_ae = z_ae.cpu().numpy()
            km = KMeans(n_clusters=n_clusters, random_state=seed, n_init=20).fit(z_ae)
            labels_ae = km.labels_
            ae_metrics.append(evaluate_clustering(z_ae, labels_ae, y))

            # GaussianMixture baseline
            gmm = GaussianMixture(n_components=n_clusters, covariance_type='diag', random_state=seed, n_init=5)
            gmm.fit(z_ae)
            labels_gmm = gmm.predict(z_ae)
            gmm_metrics.append(evaluate_clustering(z_ae, labels_gmm, y))

            # DEC baseline
            ae_pre = DeterministicAE(input_dim=dim, latent_dim=latent_dim, hidden_dims=HIDDEN_DIMS)
            ae_pre = train_ae(ae_pre, train_loader, val_loader, epochs=EPOCHS_AE, save_path=os.path.join(output_dir, f"ae_pre_k{n_clusters}_s{seed}.pth"))
            ae_pre.eval()
            ae_pre.to(DEVICE)
            ae_trained, clustering = train_dec(ae_pre, X, n_clusters=n_clusters, epochs=EPOCHS_DEC)
            with torch.no_grad():
                z_all_dec = ae_trained.encoder(torch.from_numpy(X).float().to(DEVICE)).cpu().numpy()
                q = clustering(torch.from_numpy(z_all_dec).float().to(DEVICE)).cpu().numpy()
                labels_dec = q.argmax(axis=1)
            dec_metrics.append(evaluate_clustering(z_all_dec, labels_dec, y))

            # Visualizations
            try:
                plot_umap(z_mu, labels_vade, f"VaDE n={n_clusters} seed={seed}", os.path.join(output_dir, f"umap_vade_k{n_clusters}_s{seed}.png"))
                if y is not None:
                    plot_umap(z_mu, y, f"True labels n={n_clusters} seed={seed}", os.path.join(output_dir, f"umap_true_k{n_clusters}_s{seed}.png"))
                plot_tsne(z_mu, labels_vade, f"VaDE tSNE k{n_clusters} s{seed}", os.path.join(output_dir, f"tsne_vade_k{n_clusters}_s{seed}.png"))
            except Exception as e:
                print("Visualization error:", e)

        def summarize(lst, key):
            arr = np.array([d[key] for d in lst], dtype=float)
            return float(np.nanmean(arr)), float(np.nanstd(arr))
        summary = {
            "VaDE": {m: summarize(vade_metrics, m) for m in ['silhouette','ARI','NMI']},
            "AE+KMeans": {m: summarize(ae_metrics, m) for m in ['silhouette','ARI','NMI']},
            "GMM": {m: summarize(gmm_metrics, m) for m in ['silhouette','ARI','NMI']},
            "DEC": {m: summarize(dec_metrics, m) for m in ['silhouette','ARI','NMI']}
        }
        results_summary[n_clusters] = summary
        print("\nSummary for n_clusters=", n_clusters)
        for method, metrics in summary.items():
            print(f" Method: {method}")
            for k, (mean, std) in metrics.items():
                print(f"   {k}: {mean:.4f} ± {std:.4f}")

    import json
    with open(os.path.join(output_dir, "results_summary_fixed.json"), "w") as f:
        json.dump(results_summary, f, indent=2)
    print("Saved results summary to", os.path.join(output_dir, "results_summary_fixed.json"))
    return results_summary

if __name__ == "__main__":
    set_seed(0)
    results = run_all(csv_path=CSV_PATH)
    print("Done. Outputs saved to:", OUTPUT_DIR)
```
